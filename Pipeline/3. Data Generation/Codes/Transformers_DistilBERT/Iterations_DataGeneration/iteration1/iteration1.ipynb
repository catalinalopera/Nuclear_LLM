```python
import json
from transformers import pipeline
from google.colab import files
```

```python
# Upload the file
uploaded = files.upload()
```

```python
# Assume the uploaded file is named 'input.json'
input_file = list(uploaded.keys())[0]
```

```python
# Load the dataset from the uploaded JSON file
with open(input_file, "r") as f:
    data = json.load(f)
```

```python
# Load the dataset from the uploaded JSON file
with open(input_file, "r") as f:
    data = json.load(f)
```

```python
# Load the pre-trained distilBERT model for question generation and answering
question_generator = pipeline("text2text-generation", model="mrm8488/t5-base-finetuned-question-generation-ap")
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
output_data = []
```

```python
# Generate questions and answers
for entry in data["NuclearDataset"]:
    text = entry["Text"]
    context = entry["Context"]

    # Generate question
    question = question_generator(context)[0]['generated_text']

    # Generate answer
    answer = qa_pipeline(question=question, context=context)['answer']

    output_data.append({
        "Text": text,
        "Context": context,
        "Question": question,
        "Answer": answer
    })
```

```python
# Save to a JSON file
output_file = "output.json"
with open(output_file, "w") as f:
    json.dump(output_data, f, indent=4)

print(f"Output saved to {output_file}")
```
