```python
def remove_special_characters(text):
    cleaned_text = re.sub(r'[^\w\s]', '', text)  # Remove only special characters (\w: letters and numbers, \s: spaces)
    return cleaned_text
```

```python
def clean_text(text):
    # Preservation of English letters, numbers and certain special characters only
    text = []
    for char in text:
        if char.isalnum() or char in [' ', '.', ',', '?', '!']:
            text.append(char)
        elif char == '\n':  # End-of-line characters can be preserved
            text.append(char)
        elif unicodedata.category(char)[0] == 'P':  # Punctuation marks
            text.append(' ')
    text = ''.join(text)
    return text
```
